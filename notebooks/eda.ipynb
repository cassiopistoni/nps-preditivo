```python
# ============================================================
# Análise Exploratória de Dados (EDA) - Projeto NPS Preditivo
# ============================================================

# 1. Importar bibliotecas principais
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de estilo
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
# 2. Carregar a base de dados
df = pd.read_csv("data/Base_de_dados_NPS.csv")
```

```python
# 3. Diagnóstico inicial
print("Informações gerais:")
print(df.info())
print("\nEstatísticas descritivas:")
print(df.describe(include='all'))
```

```python
# 4. Distribuição do NPS
sns.histplot(df['nps_score'], bins=10, kde=True)
plt.title("Distribuição do NPS")
plt.xlabel("NPS Score")
plt.ylabel("Frequência")
plt.show()
```

```python
# 5. Relação entre atraso de entrega e NPS
sns.boxplot(x=pd.cut(df['delivery_delay_days'], bins=[0,1,3,7,30]), y=df['nps_score'])
plt.title("NPS vs Atraso na Entrega")
plt.xlabel("Faixa de atraso (dias)")
plt.ylabel("NPS Score")
plt.show()
```

```python
# 6. Relação entre contatos no SAC e NPS
sns.boxplot(x=pd.cut(df['customer_service_contacts'], bins=[0,1,3,5,10]), y=df['nps_score'])
plt.title("NPS vs Contatos no SAC")
plt.xlabel("Número de contatos (faixas)")
plt.ylabel("NPS Score")
plt.show()
```

```python
# 7. Correlação entre variáveis numéricas
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlação")
plt.show()
```

```python
# 8. Insight inicial (exemplo)
# - Clientes com atraso > 3 dias tendem a ter NPS mais baixo.
# - Mais de 2 contatos no SAC está associado a notas menores.
# - Recompra em 30 dias é mais frequente entre clientes promotores.
```
